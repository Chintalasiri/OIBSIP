[ ] Time series analysis: plot monthly and quarterly sales trends using line charts


```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df
```

```python
df['Month'] = df['Order Date'].dt.to_period('M')
```

```python
df
```

```python
df['Quarter'] = df['Order Date'].dt.to_period('Q')
```

```python
df.columns
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales
```

plot monthly sales

```python
plt.figure(figsize=(12,5))

monthly_sales.plot(marker='o', color='blue')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)

plt.show()
```

## Observation - Monthly Sales Trend

- Monthly sales fluctuate throughout the year.
- Some months record significantly higher sales, indicating seasonal demand.
- The trend helps identify peak sales periods and months with lower business activity.

[ ] Heatmap: correlation matrix between numerical variables


```python
numerical_df = df.select_dtypes(include=['int64', 'float64'])

numerical_df.head()
```

correlation matrix

```python
corr_matrix = numerical_df.corr()

corr_matrix
```

plot heatmap

```python
plt.figure(figsize=(10,6))

sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            linewidths=0.5)

plt.title("Correlation Heatmap")

plt.show()
```

## Observation - Correlation Heatmap

- Sales and Profit have a moderate positive correlation.
- Discount shows a negative relationship with Profit.
- Most other numerical variables have weak correlations.
- The heatmap helps understand relationships between business metrics.

[ ] At least one additional visualisation of your choice that reveals a non-obvious insight

```python
region_profit = df.groupby('Region')['Profit'].sum().sort_values()

plt.figure(figsize=(8,5))

sns.barplot(x=region_profit.index,
            y=region_profit.values,
            palette='viridis')

plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")

plt.show()
```

Discount vs profit

```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=df, x='Discount', y='Profit')

plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")

plt.show()
```

## Observation - Discount vs Profit

- Higher discounts do not always increase profits.
- Several transactions with large discounts result in lower or negative profits.
- Businesses should optimize discount strategies to maintain profitability.

sales by ship mode

```python
ship_sales = df.groupby('Ship Mode')['Sales'].sum().sort_values()

plt.figure(figsize=(8,5))

sns.barplot(x=ship_sales.index, y=ship_sales.values)

plt.title("Sales by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Total Sales")

plt.xticks(rotation=20)

plt.show()
```

## Summary

This Exploratory Data Analysis (EDA) on the retail sales dataset provided valuable insights into sales trends, product performance, regional profitability, and the relationships between important business variables. The analysis helped identify top-performing products, sales patterns over time, and factors affecting profitability.

## Business Recommendations

### 1. Focus on High-Performing Products
The top-selling products contribute a significant portion of total sales. Ensure these products are always in stock and promote them through targeted marketing campaigns.

### 2. Optimize Discount Strategies
The analysis shows that higher discounts can reduce overall profit. The company should review its discount policies and offer discounts only where they improve sales without significantly affecting profitability.

### 3. Improve Performance in Low-Profit Regions
Some regions generate lower profits than others. The business should investigate the reasons behind poor performance and implement region-specific marketing and sales strategies to increase profitability.
