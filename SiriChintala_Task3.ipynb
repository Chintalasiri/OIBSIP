## Standardisation

The `Sex` column was inspected for inconsistent formatting.

The dataset already contained standardized numeric values (0 and 1) instead of text labels. Therefore, no further standardization was required for this column.

No inconsistencies were found in the categorical formatting.

```python
df.columns
```

```python
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)

IQR = Q3 - Q1

print("Q1 :", Q1)
print("Q3 :", Q3)
print("IQR :", IQR)
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit :", lower_limit)
print("Upper Limit :", upper_limit)
```

```python
outliers = df[(df["Age"] < lower_limit) | (df["Age"] > upper_limit)]

print(outliers)
```

```python
print("Number of Outliers:", len(outliers))
```
